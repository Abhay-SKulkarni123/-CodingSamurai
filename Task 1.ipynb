```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('calendar.csv')
df
```

```python
df.head()
df.info()
df.describe()
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df
```

```python
data_types_converted = df.dtypes
basic_stats_converted = df.describe(include='all')

data_types_converted
basic_stats_converted
```

```python
mean = df['price'].mean()
df['price'].fillna(mean, inplace=True)

df.drop_duplicates(inplace=True)

valid_available_values = ['t', 'f']
df = df[df['available'].isin(valid_available_values)]
df
```

```python
data_types_cleaned = df.dtypes
basic_stats_cleaned = df.describe(include='all')

data_types_cleaned
basic_stats_cleaned 
```

```python
average_price = df['price'].mean()
print(f"Average Price of Listings: ${average_price:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=30)
plt.title('Price Distribution of Airbnb Listings')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])

time_series = df.groupby('date')['price'].mean()

plt.figure(figsize=(12, 6))
time_series.plot()
plt.title('Average Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()
```

```python
df['available'] = df['available'].map({'t': 1, 'f': 0})

correlation_matrix = df.corr()

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
